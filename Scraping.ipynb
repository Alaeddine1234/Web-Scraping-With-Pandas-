2 Read .csv from Multiple URLs with Pandas  

```python
root="https://www.football-data.co.uk/mmz4281/"
```

```python
leagues=['E0', 'E1', 'E2', 'E3','EC']
frames=[]
```

```python
for league in leagues:
    df = pd.read_csv(root + '2122' + '/' + league + '.csv', encoding='unicode_escape')
    frames.append(df)
```

```python
for league in leagues:
    for season in range(15,21):
        df = pd.read_csv(root + str(season) + str(season+1) + '/' + league + '.csv', encoding='unicode_escape')
        frames.append(df)
```

```python
dict_countries = {
    'Spanish La Liga': 'SP1',
    'Spanish Segunda Division': 'SP2',
    'German Bundesliga': 'D1',
    'English Premier League': 'E0',
    'English League 1': 'E2',
    'English League 2': 'E3',
    
}
```

```python
dict_countries['Spanish La Liga']
```

```python
for league in dict_countries:
    print( dict_countries[league])
```

```python
dict_countries[league]
```

```python
dict_historical_data = {}

for league in dict_countries:
    frames = []
    for season in range(15,21):
        file_path = root + str(season) + str(season+1) + '/' + dict_countries[league] + '.csv'
        df = pd.read_csv(file_path, encoding='unicode_escape')
        df.insert(1, 'season', str(season) + str(season+1))  # mieux que juste season
        frames.append(df)
    df_concat=pd.concat(frames, ignore_index=True)
    dict_historical_data[league] = df_concat
```
